## Practice web scraping : Challenge 1

- Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page: url ='https://en.wikipedia.org/wiki/Python'

## Loading the library

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

## Storing the URL

```python
url ='https://en.wikipedia.org/wiki/Python'
```

## Getting the html code of the web page

```python
response = requests.get(url)
response.status_code
```

## Parsing the html code


```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

## Retrieving the desired info from the Soup.

```python
soup.select("#mw-content-text > div.mw-parser-output > ul > li > a")[0]["href"]
```

```python
list_link = []

for i in range(len(soup.select("#mw-content-text > div.mw-parser-output > ul > li > a"))):
    list_link.append("wikipedia.org" + soup.select("#mw-content-text > div.mw-parser-output > ul > li > a")[i]["href"])

list_link
```
